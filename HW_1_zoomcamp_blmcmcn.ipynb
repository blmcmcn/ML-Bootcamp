```python
X = df[df.Make == "Rolls-Royce"][["Engine HP", "Engine Cylinders", "highway MPG"]].drop_duplicates().values
```

```python
X
```

```python
XTX = (X.T).dot(X)
print("The inner product of matrix 'X' with itself is: \n")
XTX
```

```python
print("The inverse of 'XTX' is: \n")
np.linalg.inv(XTX)
```

```python
print("The sum of the elements of (XTX)^-1 is: \n")
np.linalg.inv(XTX).sum()
```

```python
y = [1000, 1100, 900, 1200, 1000, 850, 1300]
```

## (X<sup>T</sup>X)<sup>-1</sup>X<sup>T</sup>y = w

```python
w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
```

```python
print(f"The first weight is: {w[0]}")
```
